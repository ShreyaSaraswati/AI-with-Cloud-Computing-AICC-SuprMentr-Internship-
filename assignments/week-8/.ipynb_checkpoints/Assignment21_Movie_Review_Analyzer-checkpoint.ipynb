## Problem Statement

Build a simple sentiment analyzer and test on 5 movie reviews.

Classify reviews as:

Positive\
Negative\
Neutral

---

## Sample Reviews
Movie was amazing and story was great\
Film was boring and very slow\
Excellent acting and good direction\
Worst movie ever\
Nice visuals but bad screenplay

---

## Code

```python
import pandas as pd

reviews = [
    "movie was amazing and story was great",
    "film was boring and very slow",
    "excellent acting and good direction",
    "worst movie ever",
    "nice visuals but bad screenplay"
]

positive_words = ["good","great","excellent","amazing","nice"]
negative_words = ["bad","worst","boring","slow"]

def sentiment(text):
    
    text = text.lower()
    
    pos = sum(word in text for word in positive_words)
    
    neg = sum(word in text for word in negative_words)
    
    if pos > neg:
        return "Positive"
    
    elif neg > pos:
        return "Negative"
    
    else:
        return "Neutral"

results = [sentiment(r) for r in reviews]

df = pd.DataFrame({
    "review":reviews,
    "sentiment":results
})

print(df)
```
