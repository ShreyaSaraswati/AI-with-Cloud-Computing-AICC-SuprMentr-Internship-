## Problem Statement

Create confusing prompts to test AI behavior.

Identify how AI handles:

Conflicting instructions\
Unclear prompts\
Logical contradictions\
Example Prompts\
Explain python in 5 words but give detailed answer\
Give short answer in 200 words\
Write sentence that is positive and negative\
Tell something true and false together\
Explain like child but use technical words

---

## Code

```python
import pandas as pd

prompts = [
    "Explain python in 5 words but give detailed answer",
    "Give short answer in 200 words",
    "Write sentence that is positive and negative",
    "Tell something true and false together",
    "Explain like child but use technical words"
]

# paste actual responses from AI here
responses = [
    "Python is simple versatile language. Python is widely used...",
    
    "This is a long explanation that exceeds a short answer...",
    
    "This product is great but also terrible.",
    
    "This statement is true but also false depending on context.",
    
    "Python is like a magic tool used for computational implementation."
]

observations = [
    "AI ignored 5 word limit and gave long explanation",
    
    "Contradiction between short and long instruction",
    
    "AI produced mixed sentiment sentence",
    
    "AI tried balancing logical contradiction",
    
    "Tone is confusing (simple + technical)"
]

df = pd.DataFrame({
    "prompt":prompts,
    "AI_response":responses,
    "observation":observations
})

print(df)
```
