## Problem Statement
Suggest real-world Machine Learning problem ideas from different domains.  
For each idea, define the input (features) and expected output (prediction).

Goal: Understand how ML problems are designed by identifying features and labels.

---

## ML Problem Ideas

### 1. College – Student Performance Prediction
**Problem:** Predict whether a student might struggle in a course so teachers can provide early support.

**Input (Features):**
- Attendance percentage
- Assignment scores
- Previous exam marks
- Study hours
- Class participation

**Output (Prediction):**
- Performance category (High / Medium / At Risk)

---

### 2. Healthcare – Disease Risk Prediction
**Problem:** Use patient health data to predict the risk of disease for preventive care.

**Input (Features):**
- Age
- Weight / BMI
- Blood pressure
- Blood sugar level
- Family medical history

**Output (Prediction):**
- Disease risk level (Low / Medium / High)

---

### 3. Shopping – Product Recommendation System
**Problem:** Recommend products to customers based on their behavior and preferences.

**Input (Features):**
- Purchase history
- Browsing history
- Product ratings
- Customer preferences

**Output (Prediction):**
- Personalized product recommendations

---

### 4. Business – Product Price Prediction
**Problem:** Predict the optimal price of a product based on market trends.

**Input (Features):**
- Product category
- Brand
- Demand trends
- Competitor prices

**Output (Prediction):**
- Suggested optimal product price
