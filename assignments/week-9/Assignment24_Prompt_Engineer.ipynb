## Problem Statement

Create weak and strong prompts for:

## Sample Prompts

---

## Code

```python
import pandas as pd

tasks = ["resume","business idea","study plan"]

weak_prompts = [
    "write resume",
    "give business idea",
    "create study plan"
]

strong_prompts = [
    "create professional resume for fresher with python skills projects education contact details",
    
    "suggest low investment business idea for students include cost target customers profit",
    
    "create 4 week python study plan for beginner include daily topics practice tasks"
]

df = pd.DataFrame({
    "task":tasks,
    "weak_prompt":weak_prompts,
    "strong_prompt":strong_prompts
})

print(df)
```
