## Problem Statement

Build a simple NLP mini application to process text data and extract useful keywords using preprocessing techniques.

---

## Code

```python
import string
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')

text = """
AI is transforming the world. It helps automate tasks,
improves decision making and builds smart systems.
"""

text = text.lower()

text = text.translate(str.maketrans('', '', string.punctuation))

words = text.split()

stop_words = set(stopwords.words('english'))

keywords = [word for word in words if word not in stop_words]

print("Keywords:")
print(keywords)
```
