## Problem Statement

Load an image and analyze its numerical representation including shape, pixel values, and color channels.

---

## Code

```python
import cv2
import urllib.request
import matplotlib.pyplot as plt

# download image
url = "https://raw.githubusercontent.com/opencv/opencv/master/samples/data/lena.jpg"
urllib.request.urlretrieve(url, "sample.jpg")

# read image
image = cv2.imread("sample.jpg")

# convert BGR to RGB (important for correct color display)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# show image
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()

# image details
print("Shape (Height, Width, Channels):", image.shape)

print("Pixel value at (0,0):", image[0,0])

blue = image[:,:,0]
green = image[:,:,1]
red = image[:,:,2]

print("Blue channel shape:", blue.shape)
print("Green channel shape:", green.shape)
print("Red channel shape:", red.shape)
```
