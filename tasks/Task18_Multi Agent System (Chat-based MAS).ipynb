---

## Problem Statement
Convert the existing AI agent into a chat-based Multi-Agent System (MAS).

Add roles:

Researcher → collects information\
Writer → prepares content\
Reviewer → checks quality

Simulate conversation between agents for the topic:

"AI agents in productivity"

---

## Code 

```python
# -----------------------------
# MULTI AGENT SYSTEM (MAS)
# -----------------------------

# Agent 1 – Researcher
def researcher_agent(topic):
    
    research_data = f"""
    Research notes on {topic}:

    AI agents are software systems that perform tasks autonomously.
    
    Multi-agent systems consist of multiple AI agents working together.

    Applications include automation, cybersecurity, and productivity tools.
    """
    
    return research_data


# Agent 2 – Writer
def writer_agent(research_notes):
    
    article = f"""
    Article Draft:

    {research_notes}

    AI agents improve efficiency by automating repetitive tasks.

    Organizations use multi-agent systems to solve complex problems.
    """
    
    return article


# Agent 3 – Reviewer
def reviewer_agent(draft):
    
    reviewed = draft + """

    Review Comments:
    
    Content is clear and structured.
    
    Added clarity on real-world applications.
    """
    
    return reviewed


# -----------------------------
# CHAT BASED MAS
# -----------------------------
def multi_agent_chat(user_input):
    
    print("User:", user_input)
    
    research = researcher_agent(user_input)
    
    print("\nResearcher Agent:\n", research)
    
    
    draft = writer_agent(research)
    
    print("\nWriter Agent:\n", draft)
    
    
    final_output = reviewer_agent(draft)
    
    print("\nReviewer Agent:\n", final_output)
    
    
    return final_output


# -----------------------------
# EXECUTION
# -----------------------------
topic = "AI agents in productivity"

result = multi_agent_chat(topic)
```
