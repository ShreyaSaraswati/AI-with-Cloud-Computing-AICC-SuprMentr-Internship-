## Problem Statement
Build: "AI Resume Checker"

Features:\
User uploads resume in PDF format \
Extract text from PDF using PyPDF2\
Send resume content

---

## Code 

```python
# Imports
import io
import PyPDF2
from groq import Groq
import ipywidgets as widgets
from IPython.display import display, Markdown

# Add your Groq API Key here
GROQ_API_KEY = "your_api_key_here"

# Create Groq client
client = Groq(api_key=GROQ_API_KEY)

# Function to extract text from PDF
def extract_text_from_pdf(uploaded_file):
    text = ""
    reader = PyPDF2.PdfReader(uploaded_file)

    for page in reader.pages:
        text += page.extract_text() + "\n"

    return text


# Function to analyze resume
def analyze_resume(resume_text):

    prompt = f"""
    You are an expert ATS Resume Checker and HR Recruiter.

    Analyze the resume and provide structured output:

    1. ATS Score (out of 100)
    2. Summary of profile
    3. Strengths
    4. Weaknesses
    5. Missing keywords
    6. Skills improvement suggestions
    7. Recommended job roles
    8. Formatting improvements

    Resume:
    {resume_text}
    """

    completion = client.chat.completions.create(
        model="llama3-70b-8192",
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.3
    )

    return completion.choices[0].message.content


# File upload widget
upload = widgets.FileUpload(
    accept='.pdf',
    multiple=False,
    description="Upload Resume PDF"
)

display(upload)

# Button
button = widgets.Button(description="Analyze Resume")
display(button)

# Output area
output = widgets.Output()
display(output)


# Button click action
def on_button_click(b):

    with output:
        output.clear_output()

        if upload.value:

            file_data = list(upload.value.values())[0]
            content = file_data['content']

            pdf_file = io.BytesIO(content)

            print("Extracting text from PDF...\n")

            resume_text = extract_text_from_pdf(pdf_file)

            print("Analyzing resume using AI...\n")

            result = analyze_resume(resume_text)

            display(Markdown("## AI Resume Feedback"))
            print(result)

        else:
            print("Please upload a PDF file first.")


button.on_click(on_button_click)
```
