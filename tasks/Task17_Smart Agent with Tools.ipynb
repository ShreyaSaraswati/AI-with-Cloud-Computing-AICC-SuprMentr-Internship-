## Problem Statement
Extend the current AI agent by adding tools:

Weather API tool\
File reader tool

Agent should process the input:

"check weather in bangalore"

---

## Code 

```python
import requests

# ----------------------
# TOOL 1 – Weather Tool
# ----------------------
def weather_tool(city):
    
    # free public weather API
    url = f"https://wttr.in/{city}?format=j1"
    
    response = requests.get(url)
    
    data = response.json()
    
    temp = data["current_condition"][0]["temp_C"]
    
    description = data["current_condition"][0]["weatherDesc"][0]["value"]
    
    result = f"Weather in {city}: {temp}°C, {description}"
    
    return result


# ----------------------
# TOOL 2 – File Reader
# ----------------------
def file_reader_tool(filename):
    
    try:
        
        with open(filename,"r") as f:
            
            content = f.read()
        
        return content
    
    except:
        
        return "file not found"


# ----------------------
# AGENT
# ----------------------
def smart_agent(user_input):
    
    user_input = user_input.lower()
    
    
    # planning
    if "weather" in user_input:
        
        city = user_input.split("in")[-1].strip()
        
        return weather_tool(city)
    
    
    elif "read file" in user_input:
        
        filename = user_input.split("file")[-1].strip()
        
        return file_reader_tool(filename)
    
    
    else:
        
        return "task not supported"


# ----------------------
# EXECUTION
# ----------------------
query = "check weather in bangalore"

output = smart_agent(query)

print("User Input:", query)

print("Agent Output:", output)
```
