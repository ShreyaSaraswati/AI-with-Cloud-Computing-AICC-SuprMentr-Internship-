## Problem Statement
Build: "AI Study Assistant"\
Features:
- User enters topic
- API returns explanation 

---

## Code 

```python
from openai import OpenAI

# create client
key = input("Enter API key: ")
client = OpenAI(api_key=key)

print("AI Study Assistant (type 'exit' to stop)")

while True:
    
    topic = input("\nEnter topic: ")

    if topic.lower() == "exit":
        break

    response = client.responses.create(
        model="gpt-4.1-mini",
        input=f"""
Explain the topic in simple student-friendly way.

Give:
1. Definition
2. Example
3. Short summary

Topic: {topic}
"""
    )

    print("\nExplanation:\n")
    print(response.output_text)
```
